```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# HEROIN

```python
OH_Heroin = pd.read_excel("Data/OH_Heroin.xlsx")
```

```python
date = [2010,2011,2012,2013,2014,2015,2016,2017]
HEROIN_LORAIN = [14,21,11,31,42,476,467,353]
HEROIN_LAKE = [193,397,489,811,721,1065,942,568]
HEROIN_OTTAWA = [4,19,12,26,27,23,50,21]
```

## LORAIN和LAKE的关系图

```python
plt.figure(figsize=(9,5))
plt.plot(date,HEROIN_LORAIN,'r',date,HEROIN_LAKE,'b')
plt.title('Heroin Drug Reports chart')
plt.xlabel('Year')
plt.ylabel('TotalDrugReportsCounty')
plt.legend(['LORAIN','LAKE'],prop={'family':'SimHei','size':14})
plt.savefig('Heroin_LORAIN_LAKE.jpg')
```

## LORAIN和LAKE的相关系数（暂时不用管）

```python
a = np.array(HEROIN_LORAIN)
b = np.array(HEROIN_LAKE)
data = np.vstack((a,b)).T
DF = pd.DataFrame(data,columns=['O_LORAIN','O_LAKE'])
DF.corr() 
```
